```python
import pandas as pd
import glob
```

```python
path = r'./raw_data/cred_2018/' # use your path
files = glob.glob(path + "credores_empenhos_2018_p1.csv")

credores_list = []

for filename in files:
    df = pd.read_csv(filename, index_col=None, header=0)
    credores_list.append(df)

cred_emp18 = pd.concat(credores_list, axis=0, ignore_index=True, sort=True).reset_index(drop=True)
cred_emp18 = cred_emp18.drop(cred_emp18.filter(like=r'Unnamed').columns, axis=1)
```
